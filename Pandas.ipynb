#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
!wget https://github.com/hse-ds/iad-intro-ds/raw/master/2023/homeworks/Data.zip
!unzip /content/Data.zip # разархивируем таблицы из архив в папку Data
```

```python
groups = [] # создаем пустой список, в который будем добавлять таблицы, содержащие полную информацию про каждую группу
for index in range(0, 10): # цикл по всем индексам таблиц от 0 до 9
  info_name = "Data/Students_info_" + str(index) + ".csv" # название текущей таблицы про студентов
  marks_name = "Data/Students_marks_" + str(index) + ".csv" # название текущей таблицы про оценки
  df_info = pd.read_csv(info_name) # читаем текущую таблицу про студентов
  df_marks = pd.read_csv(marks_name) # читаем текущую таблицу про оценки
  df_group = pd.merge(df_info, df_marks, on="index", how="left") # соединяем таблицы в одну, left-join
  groups.append(df_group) # добавляем таблицу в список
df = pd.concat(groups) # соединяем все таблицы в одну
df.head(10) # выводим первые 10 строк итоговой таблицы
```

    * Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?
    * Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?

```python
# Какая доля студентов сдала экзамен по математике (passmark >= 50)?
np.sum(df["math score"] >= passmark) / (df.shape[0]) # считаем количество студентов, которые сдали экзамен по математике и делим их на общее количество студентов
```

```python
# Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?
completed = np.sum((df["test preparation"] == "completed") & (df["math score"] >= passmark)) # считаем количество студентов, которые проходили курс подготовки и сдали экзамен
all_tested = np.sum((df["test preparation"] == "completed")) # считаем количество студентов, которые проходили курс подготовки
completed / all_tested # считаем долю студентов, проходивших курс подготовки к экзамену и сдавших его
```

```python
# Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?
not_completed = np.sum((df["gender"] == "female") & (df["test preparation"] == "none") & (df["math score"] < passmark)) # считаем количество женщин, которые не проходили курс подготовки и не сдали экзамен
all_not_tested= np.sum((df["gender"] == "female") & (df["test preparation"] == "none")) # считаем количество женщин, которые не проходили курс подготовки
not_completed / all_not_tested # считаем долю женщин, не проходивших курс подготовки к экзамену и не сдавших его
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def math_function(line): # функция для определения значения в столбец по математике
  if line['math score'] >= passmark: # если значение по математике выше проходной, то возвращаем P
        return "P"
  else: # иначе возвращаем F
    return "F"
def reading_function(line): # функция для определения значения в столбец по чтению
  if line['reading score'] >= passmark: # если значение по математике выше проходной, то возвращаем P
        return "P"
  else:
    return "F"
def writing_function(line): # функция для определения значения в столбец по письму
  if line['writing score'] >= passmark: # если значение по математике выше проходной, то возвращаем P
        return "P"
  else: # иначе возвращаем F
    return "F"

df['Math_PassStatus'] = df.apply(math_function, axis=1) # в новый столбец записываем значение согласно функции
print("Экзамен по математике") # вспомогательный вывод
print(df["Math_PassStatus"].value_counts()) # считаем количество студентов, сдавших и не сдавших экзамен по математике
df['Reading_PassStatus'] = df.apply(reading_function, axis=1) # в новый столбец записываем значение согласно функции
print("Экзамен по чтению") # вспомогательный вывод
print(df["Reading_PassStatus"].value_counts()) # считаем количество студентов, сдавших и не сдавших экзамен по чтению
df['Writing_PassStatus'] = df.apply(writing_function, axis=1) # в новый столбец записываем значение согласно функции
print("Экзамен по письму") # вспомогательный вывод
print(df["Writing_PassStatus"].value_counts()) # считаем количество студентов, сдавших и не сдавших экзамен по письму
df.head(10) # выводим первые 10 строк таблицы

```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def function(line):
  if line['Math_PassStatus'] != "P" or line['Reading_PassStatus'] != "P" or line['Writing_PassStatus'] != "P": # если хотя бы один экзамен не сдан, то возвращаем F
        return "F"
  else: # иначе возвращаем P
    return "P"

df["OverAll_PassStatus"] = df.apply(function, axis=1) # создаем новый стоблец, который показывает сдал студент все экзамены или нет, значение определяется согласно функции
print("Количество студентов, сдавших все экзамены равно", end=' ') # вспомогательный вывод
print(df["OverAll_PassStatus"].value_counts()['P']) # считаем количество студентов, сдавших все экзамены
df.head(10) # выводим первые 10 строк таблицы
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
  if(average_mark > 90): # если средний балл по всем предметам больше 90, то возвращаем A
    return "A"
  elif(average_mark > 80 and average_mark <= 90): # если средний балл по всем предметам больше 80, но меньше или равен 90, то возвращаем B
    return "B"
  elif(average_mark > 70 and average_mark <= 80): # если средний балл по всем предметам больше 70, но меньше или равен 80, то возвращаем C
    return "C"
  elif(average_mark > 60 and average_mark <= 70): # если средний балл по всем предметам больше 60, но меньше или равен 70, то возвращаем D
    return "D"
  elif(average_mark >= 50 and average_mark <= 60): # если средний балл по всем предметам больше или равен 50, но меньше или равен 60, то возвращаем B
    return "B"
  elif(average_mark < 50): # если средний балл по всем предметам меньше 50, то возвращаем F
    return "F"

df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade) # добавляем в таблицу новый столбец, значение которого определяется функцией,
                                                                                                # которая принимает средний балл по всем предметам
print("Оценка и количество людей, которые её получили") # вспомогательный вывод
print(df['Grade'].value_counts()) # выводим количество студентов, получивших каждую оценку
df.head(10) # выводим первые 10 строк таблиц
```
